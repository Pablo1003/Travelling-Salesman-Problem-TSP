Haremos una implementación funcional, aunque sería más apropiado definir una clase

```python
import random
import copy

def create (alphabet, N=100): # crea población
    return [random.sample(alphabet, len(alphabet)) for _ in range(N)]

def sort_pop (pop, fit): # devuelve una tupla: la población ordenada por fitness y la lista de fitness ordenada
    sorted_pop = sorted(pop, key=fit)
    sorted_fit = [fit(x) for x in sorted_pop]
    return sorted_pop, sorted_fit

def select (pop, T): # devuelve un individuo seleccionado por torneo, devuelve una copia para evitar efectos laterales
    index_list = [random.randrange(len(pop)) for _ in range(T)]
    return copy.deepcopy(pop[min(index_list)])

def crossover (ind1, ind2, pcross):
    # devuelve una tupla con los dos hijos
    if pcross < random.random():
        return (ind1, ind2)

    mask = random.sample((0,1), k=len(ind1), counts=(len(ind1), len(ind1)))
    hijo1 = [None]*len(ind1)
    hijo2 = [None]*len(ind1)
    for i in range(len(ind1)):
        if mask[i] == 0:
            hijo1[i]=ind1[i]
            hijo2[i]=ind2[i]

    for i in range(len(ind1)):
        if ind2[i] not in hijo1:
            # busca el primer hueco
            for j in range(len(ind1)):
                if hijo1[j] == None:
                    hijo1[j] = ind2[i]
                    break
        if ind1[i] not in hijo2:
            # busca el primer hueco
            for j in range(len(ind1)):
                if hijo2[j] == None:
                    hijo2[j] = ind1[i]
                    break

    return (hijo1, hijo2)


def mutate (ind, pmut): # la mutación consistirá en intercambiar elementos
    # devuelve el individuo mutado
    if pmut < random.random():
        return ind

    gen1 = random.randrange(len(ind))
    gen2 = random.randrange(len(ind))

    ind[gen1], ind[gen2] = ind[gen2], ind[gen1]

    return ind


def evolve (pop, fit, pmut, pcross=0.7, ngen=100, T=2, trace=0, elitism=False, maximize=False):
    # devuelve la población evolucionada
    # si trace > 0, muestra información cada trace generaciones
    # si elitism=True, se aplica elitismo
    # si maximize=True, se busca el máximo en lugar del mínimo
    best = None
    best_fit = None
    pop_size = len(pop)
    for gen in range(ngen):
        pmut_gen = max(pmut*(1-gen/ngen), 0.5) # reducimos la probabilidad de mutación
        newpop = []
        for i in range(pop_size//2):
            # seleccionamos dos individuos
            ind1 = select(pop, T)
            ind2 = select(pop, T)
            # los cruzamos
            hijo1, hijo2 = crossover(ind1, ind2, pcross)
            # los mutamos
            hijo1 = mutate(hijo1, pmut_gen, gen)
            hijo2 = mutate(hijo2, pmut_gen, gen)
            # los añadimos a la nueva población
            newpop.append(hijo1)
            newpop.append(hijo2)
        # elitismo
        if elitism and best != None:
            # eliminamos un individuo aleatorio
            newpop.pop(random.randrange(len(newpop)))
            # añadimos el mejor de la generación anterior
            newpop.append(best)

        # evaluamos la nueva población
        pop, fit_pop = sort_pop(newpop, fit)
        if maximize:
            # damos la vuelta a las listas
            pop.reverse()
            fit_pop.reverse()
        # nos quedamos con el mejor
        if maximize:
            if best_fit == None or fit_pop[0] > best_fit:
                best = pop[0]
                best_fit = fit_pop[0]
        else:
            if best_fit == None or fit_pop[0] < best_fit:
                best = pop[0]
                best_fit = fit_pop[0]
        # mostramos información
        if trace > 0 and gen % trace == 0:
            print("Generación: ", gen)
            print("Mejor: ", best)
            print("Fitness: ", best_fit)
    return pop
```

```python
# pero el azar siempre es nuestro aliado
# el uso de varios operadores de mutación ayuda a salir de mínimos locales

def mutate_swap (ind, pmut): # la mutación consistirá en intercambiar elementos
    # devuelve el individuo mutado
    if pmut < random.random():
        return ind

    gen1 = random.randrange(len(ind))
    gen2 = random.randrange(len(ind))

    ind[gen1], ind[gen2] = ind[gen2], ind[gen1]

    return ind

def mutate_insert (ind, pmut): # la mutación consistirá en borrar un elemento e insertarlo en otro lado
    # devuelve el individuo mutado
    if pmut < random.random():
        return ind

    gen1 = random.randrange(len(ind))
    pos = random.randrange(len(ind))

    ind.insert(pos, ind.pop(gen1))

    return ind

def mutate_reverse (ind, pmut): # selecciona dos puntos e invierte la secuencia entre ellos
    # devuelve el individuo mutado
    if pmut < random.random():
        return ind

    gen1 = random.randrange(len(ind))
    gen2 = random.randrange(len(ind))

    if gen1 > gen2:
        gen1, gen2 = gen2, gen1

    ind[gen1:gen2] = reversed(ind[gen1:gen2])

    return ind

def mutate_reverse_out (ind, pmut): # selecciona dos puntos e intercambia los extremos de lado
    # devuelve el individuo mutado
    if pmut < random.random():
        return ind

    gen1 = random.randrange(len(ind))
    gen2 = random.randrange(len(ind))

    if gen1 > gen2:
        gen1, gen2 = gen2, gen1

    mid = ind[gen1:gen2]
    ini = ind[:gen1]
    fin = ind[gen2:]
    ind = fin + mid + ini

    return ind

def mutate_reverse_mini (ind, pmut): # selecciona dos puntos e invierte la secuencia entre ellos
    # devuelve el individuo mutado
    if pmut < random.random():
        return ind

    gen1 = random.randrange(len(ind))
    gen2 = gen1 + random.randrange(1, 4)

    ind[gen1:gen2] = reversed(ind[gen1:gen2])

    return ind

mutations = [mutate_swap, mutate_insert, mutate_reverse, mutate_reverse_out]#
mutations2 = [mutate_swap, mutate_insert, mutate_reverse_mini]

def mutate (ind, pmut, gen=0):
    if gen < 700:
        operator = random.choice(mutations)
    else:
        operator = random.choice(mutations2)
    return operator(ind, pmut)
```

```python
N = 100
best_fit = 1e5
best = None
pops = []
for i in range(20):
    pop = create(cities, N)
    pop = evolve(pop, globaldist, 75/100, ngen=1000, trace=100, T=6, pcross=0.0, elitism=True)
    for e in pop:
        pops.append(globaldist(e))
    if globaldist(pop[0]) < best_fit:
        best_fit = globaldist(pop[0])
        best = pop[0]
mean = np.mean(np.asarray(pops))
std = np.std(np.asarray(pops))
print("Mean: ", mean)
print("Std: ", std)
print("Best: ", best_fit)
draw(best)
```

```python
# Función himmelblau
# si se pone onlyone a True

onlyone = False
def himmelblau (ch):
	x = ch[0]
	y = ch[1]
	fxy = (x**2 + y - 11)**2 + (x + y**2 -7)**2
	if onlyone and (x>0 or y>0): # himmelblau modified to have only one global minimum (-3.77, -3.28)
		fxy += 0.5
	return fxy
```

```python
def evolve (pop, fit, pmut, pcross=0.7, ngen=100, T=2, trace=0, elitism=False, maximize=False):
    # devuelve la población evolucionada
    # si trace > 0, muestra información cada trace generaciones
    # si elitism=True, se aplica elitismo
    # si maximize=True, se busca el máximo en lugar del mínimo
    best = None
    best_fit = None
    pop_size = len(pop)
    for gen in range(ngen):
        newpop = []
        for i in range(pop_size//2):
            # seleccionamos dos individuos
            ind1 = select(pop, T)
            ind2 = select(pop, T)
            # los cruzamos
            hijo1, hijo2 = crossover(ind1, ind2, pcross)
            # los mutamos
            hijo1 = mutate(hijo1, pmut)
            hijo2 = mutate(hijo2, pmut)
            # los añadimos a la nueva población
            newpop.append(hijo1)
            newpop.append(hijo2)
        # elitismo
        if elitism and best != None:
            # eliminamos un individuo aleatorio
            newpop.pop(random.randrange(len(newpop)))
            # añadimos el mejor de la generación anterior
            newpop.append(best)

        # evaluamos la nueva población
        pop, fit_pop = sort_pop(newpop, fit)
        if maximize:
            pop.reverse()
            fit_pop.reverse()
        # nos quedamos con el mejor
        if maximize:
            if best_fit == None or fit_pop[0] > best_fit:
                best = pop[0]
                best_fit = fit_pop[0]
        else:
            if best_fit == None or fit_pop[0] < best_fit:
                best = pop[0]
                best_fit = fit_pop[0]
        # mostramos información
        if trace > 0 and gen % trace == 0:
            print("Generación: ", gen)
            print("Mejor: ", best)
            print("Fitness: ", best_fit)
    return pop
```

```python
# Define operadores de números reales

rang = (-5, 5) # al no hacerlo con clases, debemos definir el rango como variable global

def create (alphabet=None, N=50): # crea y devuelve población; alphabet aquí no se usa
    return [(random.uniform(rang[0], rang[1]), random.uniform(rang[0], rang[1])) for _ in range(N)]

def crossover (ind1, ind2, pcross): # devuelve el cruce (emparejamiento) de dos individuos
    if pcross < random.random():
        return (ind1, ind2)

    # se aplica cruce aritmético
    beta = random.random()
    hijo1 = (beta*ind1[0] + (1-beta)*ind2[0], beta*ind1[1] + (1-beta)*ind2[1])
    hijo2 = (beta*ind2[0] + (1-beta)*ind1[0], beta*ind2[1] + (1-beta)*ind1[1])

    return (hijo1, hijo2)

def mutate (ind, pmut): # devuelve individuo mutado;
    if pmut < random.random():
        return ind

    ind = (ind[0] + random.uniform(rang[0], rang[1])/10, ind[1] + random.uniform(rang[0], rang[1])/10)
    #ind = (random.uniform(rang[0], rang[1]), random.uniform(rang[0], rang[1]))
    # clip to range
    ind = (max(min(ind[0], rang[1]), rang[0]), max(min(ind[1], rang[1]), rang[0]))

    return ind
```

```python
# fitness para himmelblau: valor mínimo de la función

def fitness_himmel (ch):
	return 1 / (1 + himmelblau(ch))
```

```python
N = 100
best_fit = 1e5
pops = []
for i in range(20):
    pop = create(N=50)
    pop = evolve(pop, fitness_himmel, 20/100, ngen=70, trace=0, T=4, pcross=0.6, maximize=True, elitism=True)
    for e in pop:
        pops.append(fitness_himmel(e))
    if fitness_himmel(pop[0]) < best_fit:
        best_fit = fitness_himmel(pop[0])
mean = np.mean(np.asarray(pops))
std = np.std(np.asarray(pops))
print("Mean: ", mean)
print("Std: ", std)
print("Best: ", best_fit)
```
